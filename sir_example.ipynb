## SIR example

explain about one SIR simulation

```python
S0 = 0.99 # 999 people suscetible to the disease
I0 = 0.01 # 1 infected person
R0 = 0.0   # 0 recovered people

t, S, I, R = simulate_sir_epidemics(
    infected_people_per_day = 0.5, 
    infection_duration = 4, 
    days_of_simulation = 30, 
    initial_state = (S0, I0, R0))

plot_sir(t, S, I, R)
```

```python
# Covid-19

S0 = 0.999 # 999 people suscetible to the disease
I0 = 0.001 # 1 infected person
R0 = 0.0   # 0 recovered people

t, S, I, R = simulate_sir_epidemics(
    infected_people_per_day = 2, 
    infection_duration = 15, 
    days_of_simulation = 50, 
    initial_state = (S0, I0, R0))

plot_sir(t, S, I, R)
```

```python
t, S, I, R = simulate_sir_epidemics(
    infected_people_per_day = 0.5, 
    infection_duration = 4, 
    days_of_simulation = 2, 
    initial_state = (S0, I0, R0))

print(t)
print(S)
print(I)
print(R)
```

```python
St, It, Rt = S0, I0, R0
dSdt, dIdt, dRdt = compartments_derivative((S0, I0, R0), 4, 0.25)

print( St + dSdt, It + dIdt, Rt + dRdt )
```
