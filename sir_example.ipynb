## SIR example

explain about one SIR simulation

```python
S0 = 0.99 # 999 people suscetible to the disease
I0 = 0.01 # 1 infected person
R0 = 0.0   # 0 recovered people

initial_state = (S0, I0, R0)

def simulate_and_plot_sir_epidemics(infected_people_per_day, infection_duration, days_of_simulation, initial_state):
    t, S, I, R = simulate_sir_epidemics(infected_people_per_day, infection_duration, days_of_simulation, initial_state)
    plot_sir(t, S, I, R)
```

```python
# Flu
simulate_and_plot_sir_epidemics(
    infected_people_per_day = 1, 
    infection_duration = 4, 
    days_of_simulation = 50, 
    initial_state = initial_state
)
```

```python
# Flu (with controlled contacts)
simulate_and_plot_sir_epidemics(
    infected_people_per_day = 0.5, 
    infection_duration = 4, 
    days_of_simulation = 50, 
    initial_state = initial_state
)
```

```python
# Covid-19 (one person infects four people each day)
simulate_and_plot_sir_epidemics(
    infected_people_per_day = 4, 
    infection_duration = 15, 
    days_of_simulation = 50, 
    initial_state = initial_state
)
```

```python
# Covid-19 (one person infects two people each day)
simulate_and_plot_sir_epidemics(
    infected_people_per_day = 2, 
    infection_duration = 15, 
    days_of_simulation = 50, 
    initial_state = initial_state
)
```

```python
# Covid-19 (one person infects another one each day)
simulate_and_plot_sir_epidemics(
    infected_people_per_day = 1, 
    infection_duration = 15, 
    days_of_simulation = 50, 
    initial_state = initial_state
)
```
